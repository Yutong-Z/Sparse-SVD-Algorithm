# Lung cancer data

```python
import numpy as np
import pandas as pd
import seaborn as sns
from SSVD663 import SSVD
from SSVD663 import SSVD_multi
```

```python
df = pd.read_csv('./Data/LungCancerData.csv', sep=' ', header = None).T
cancer = np.array(df)
```

```python
u, s, v = SSVD(cancer)
cancer_new = u @ s @ np.sort(np.ravel(v))[::-1].reshape(1,-1)
sns.heatmap(cancer_new, vmin=-1, vmax=1, cmap='seismic')
```

```python
u, s, v = SSVD_multi(cancer)
cancer_new = u @ s @ np.sort(np.ravel(v))[::-1].reshape(1,-1)
sns.heatmap(cancer_new, vmin=-1, vmax=1, cmap='seismic')
```
